```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Data Collection and Preprocessing

```python
df = pd.read_csv('/content/17) mail_data.csv')
df.head()
```

```python
# replace the null values with a null string

#df = df.where((pd.notnull(df)), '')
```

```python
df.isnull().sum()
```

```python
df.shape
```

Label Encording

spam --> 0

ham --> 1

```python
df.loc[df['Category'] == 'spam', 'Category'] = 0
df.loc[df['Category'] == 'ham', 'Category'] = 1
```

```python
#df.replace({'Category': {'spam':0, 'ham':1}}, inplace=True)
df.head()
```

```python
X = df.drop('Category', axis=1)
y = df.Category
```

Spliting the data in to train and test data

```python
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.2, random_state=3)
```

```python
X.shape, X_train.shape, X_test.shape
```

Feature Extraction

```python
# transform text to feature vectors

# min_df      = if a word is not use more than once we reject it
# stop_words  = (did, is, the ... etc) ignored from data set
# lowercase   = convert all to lower case

feature_extraction = TfidfVectorizer(min_df = 1, stop_words = 'english', lowercase=True)


X_train_features = feature_extraction.fit_transform(X_train)
X_test_features = feature_extraction.transform(X_test)
```

```python
X_train_features.toarray()
```

```python
X_train_features, X_test_features
```

```python
# convert y_train , y_test to integers

y_train = y_train.astype('int')
y_test = y_test.astype('int')
```

```python
X_train_features.shape
```

```python
X_train.head()
```

```python
y_train.shape
```

```python
y_train
```

Training Model

```python
model = LogisticRegression()
```

```python
# training Logistic Regression with training data

model.fit(X_train_features, y_train)
```

Evaluating trained model

```python
# Prediction on training data

prediction_on_training = model.predict(X_train_features)
accuracy = accuracy_score(y_train, prediction_on_training)
accuracy
```

```python
# Prediction on testing data

prediction_on_testing = model.predict(X_test_features)
accuracy = accuracy_score(y_test, prediction_on_testing)
accuracy
```

Build Predictive system

```python
input_mail = ["I've been searching for the right words to thank you for this breather. I promise i wont take your help for granted and will fulfil my promise. You have been wonderful and a blessing at all times."]

#convert to features

input_features = feature_extraction.transform(input_mail)
input_features
```
