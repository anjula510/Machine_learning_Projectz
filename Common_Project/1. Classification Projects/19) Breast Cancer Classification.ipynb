```python
import numpy as np
import pandas as pd
import sklearn.datasets

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Data Collection and Processing

```python
df1 = sklearn.datasets.load_breast_cancer()
```

```python
# convert data to pandas DataFrame (Array to table)

df = pd.DataFrame(df1.data, columns=df1.feature_names)
df
```

```python
# adding target column

df['label'] = df1.target

df.head()
```

```python
df.shape
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
#distribution of target variable

df['label'].value_counts()
```

Benign => 1

Malignant => 0

```python
df.groupby('label').mean()
```

seperate Features and target

```python
X = df.drop('label', axis=1)
y = df.label 
```

Splitting data in to train and test data

```python
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.2, random_state=2)
```

```python
X.shape, X_train.shape, X_test.shape
```

Model fitting

```python
model = LogisticRegression()

model.fit(X_train, y_train)
```

Evolution 

Training Prediction

```python
training_prediction = model.predict(X_train)
training_prediction
```

```python
Accuracy_score = accuracy_score(y_train, training_prediction)
Accuracy_score
```

Testing data Prediction

```python
testing_prediction = model.predict(X_test)
Accuracy_prediction = accuracy_score(y_test, testing_prediction)
Accuracy_prediction
```
